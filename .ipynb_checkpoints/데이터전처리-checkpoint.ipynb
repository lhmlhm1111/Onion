```python
import pandas as pd
```

```python
data=pd.read_excel("(합침2020-06-10 002409)17_8_14-20.xlsx")
data.head()
```

```python
a='\n1.본 자료는 출하 또는 구매시 참고로 사용할 수도 있지만, 계약•납품(산출기준자료), 사실조회(손해배상청구 등)등을 위한\n  근거•증빙자료로 사용할 수 없음을 알려드립니다.\n2.과일류는 경매시간으로 인해 오전 11시 이후에 변경될 수 있습니다.(명절 성수기 반입량 증가 시 오후 2시 이후 확정될 \n  수도 있습니다.)\n3.도매시장법인거래정보>품목별가격은 도매법인의 경매(정가수의) 가격이며, 중도매인직접거래(상장예외) 품목의 \n  판매가격은 중도매인직접거래정보>품목별가격정보를 참고하시기 바랍니다.\n4.출하자, 산지 및 품종에 따라 실제 거래가격과 차이가 날 수 있으며, 일부 품목의 경우 계절•품종별 요인에 의해 실제 \n  중량과 차이가 날 수 있습니다.\n5. 상장거래 품목별 가격 등급은 “경매(정가수의) 가격”을 경매 물량에 따라 일정 비율로 강제 배분하여 산출한 등급으로,\n  산지에서 부여한 등급(실시간 거래정보-실시간거래현황·결과)과는 무관합니다.\n(품목, 품종별로 비율이 다르며 보편적으로 특[5%],상[25%],보통[40%],하[30%]\n※ 전일대비(%): 100일 경우 전일과 동일한 가격, 110일 경우 10% 상승한 가격,90일 경우 10% 하락한 가격입니다.'
b= "[ 가락시장 ]"
c = "날짜"
```

```python
data=data.iloc[6:]
data=data[data["Unnamed: 0"] != a]
data=data[data["Unnamed: 0"] != b]
data=data[data["Unnamed: 0"].isnull() == False]
data=data[data["Unnamed: 0"] != c]
data=data.dropna(axis=1)
```

```python
columns_dict = {'Unnamed: 0':"날짜", 'Unnamed: 4':"품목", 'Unnamed: 7':"품질",
               'Unnamed: 9':"거래단위", 'Unnamed: 12':"최저가", 'Unnamed: 13':"최고가",
               "Unnamed: 14":"평균가", "Unnamed: 16":"전일평균대비등락", 
                "Unnamed: 17":"전7일평균대비등락", "Unnamed: 19":"전년동월동일평균대비등락"}
```

```python
data=data.rename(columns=columns_dict)
```

```python
data=data.reset_index().drop("index", axis=1)
```

```python
data
```

```python
f = lambda x : x.replace("\n", "")
f2 = lambda x : int(x.replace(",", ""))
```

```python
data["전7일평균대비등락"]=data["전7일평균대비등락"].apply(f)
data["전년동월동일평균대비등락"]=data["전년동월동일평균대비등락"].apply(f)
data["최저가"] = data["최저가"].apply(f2)
data["최고가"] = data["최고가"].apply(f2)
data["평균가"] = data["평균가"].apply(f2)
data["전일평균대비등락"] = data["전일평균대비등락"].apply(f2)
```

```python
data["날짜"]=data["날짜"].apply(lambda x : pd.to_datetime(x))
```

```python
data.to_csv("201706-202006_양파가격.csv", encoding="utf-8-sig")
```
