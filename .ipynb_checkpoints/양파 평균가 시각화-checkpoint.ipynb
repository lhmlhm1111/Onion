```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("201706-202006_양파가격.csv")
df=df.drop("Unnamed: 0", axis=1)
df.head()
```

```python
df["날짜"] = df.날짜.apply(lambda x : pd.to_datetime(x))
```

```python
df["연도"]=df.날짜.dt.year
```

```python
df["품목_종합"] = df["품목"] + "_" + df["품질"] + "_" + df["거래단위"]
```

```python
df["월일"]=df["날짜"].apply(lambda x : pd.to_datetime(x).strftime('%m-%d'))
```

```python
df.품목_종합.unique()
```

```python
df_양파_특_1 = df[df["품목_종합"] == "양파_특_1kg"]
```

```python
df_양파_특_1=df_양파_특_1.sort_values("월일")
```

```python
plt.style.use("ggplot")
plt.rcParams['font.family'] = "NanumGothic" 
plt.figure(figsize=[20,10])
plt.xticks(fontsize=10,rotation=90)
sns.scatterplot(x=df_양파_특_1.월일, y=df_양파_특_1.평균가, hue=df_양파_특_1.연도)
```
